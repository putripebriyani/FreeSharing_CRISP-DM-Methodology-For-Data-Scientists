# 1. **Import** Library yang dibutuhkan
Matplotlib, yaitu library yang digunakan untuk membuat grafik plot sesuai kebutuhan

Numpy, digunakan untuk kebutuhan scientific

Pandas, yaitu library yang digunakan untuk memanipulasi data seperti membuat tabel, mengubah dimensi data, mencetak data dan sebagainya

Sklearn, adalah library untuk berbagi metode dan algoritma yang digunakan dalam machine learning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

# 2. Menyiapkan **Dataset**

```python
#Menyiapkan data dan memanggil dataset
Dataset = pd.read_csv('/content/Data.csv')
Dataset.keys()
```

# 3. Menampilkan dataset

```python
dataku = pd.DataFrame(Dataset)
dataku.head()
```

# **4**. Konversi ke data array

```python
X = np.asarray(Dataset)
print (X)
```

```python
#Menampilkan data dalam bentuk scatter plot
# Ubah X menjadi DataFrame dan pisahkan kolom berdasarkan pemisah titik koma
import pandas as pd # mengimport library pandas
import numpy as np # mengimport library numpy
import matplotlib.pyplot as plt # mengimport library matplotlib

X = pd.DataFrame(X)[0].str.split(';', expand=True).astype(float).to_numpy()
# mengubah array X menjadi DataFrame
# memilih kolom pertama [0] dan memisahkan string berdasarkan titik koma (;)
# mengubah tipe data menjadi float
# mengubah kembali menjadi array numpy

plt.scatter(X[:,0],X[:,1], label='True Position') # membuat scatter plot dengan kolom pertama sebagai sumbu x dan kolom kedua sebagai sumbu y
plt.xlabel("Gaji") # memberi label pada sumbu x
plt.ylabel("Pengeluaran") # memberi label pada sumbu y
plt.title("Grafik konsumen") # memberi judul pada grafik
plt.show() # menampilkan grafik
```

# 5. MENGGUNAKAN LIBRARY K-MEANS

```python
# Mengaktifkan K Means dengan jumlah K = 2
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
# Menampilkan nilai centroid
# yang di generate oleh algoritma

print(kmeans.cluster_centers_)
```

```python
# Menampilkan label data point
print(kmeans.labels_)
```

# 6. MENAMPILKAN OUTPUT

```python
# Plot data point
# Memvisualisasikan bagaimana telah di klasterisasikan

plt.scatter(X[:,0],X[:,1], c=kmeans.labels_, cmap='rainbow')
plt.xlabel("Gaji")
plt.ylabel("Pengeluaran")
plt.title("Grafik Konsumen")
plt.show
```

```python
# Plot data point
# Memvisualisasikan bagaimana telah di klasterisasikan
# Menampilkan centroids dengan warna hitam
plt.scatter(X[:,0],X[:,1], c=kmeans.labels_,
		cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0],
		kmeans.cluster_centers_[:,1],
		color='black')
plt.xlabel("Gaji")
plt.ylabel("Pengeluaran")
plt.title("Grafik Konsumen")
plt.show()
```
